```python
import matplotlib.pyplot as plt
```

```python
# plt.scatter(x, y, color="color_name", marker="marker style", label="label_name")
hours_studied = [1, 2, 3, 4, 5, 6, 7, 8]
exam_scores = [50, 55, 60, 65, 70, 75, 80, 85]

# scatter plot
plt.scatter(hours_studied, exam_scores, color="green", marker="o", label="Studnt Data")

plt.xlabel("Hours Studied")
plt.ylabel("Exam Score")
plt.title("Relationship Between Study Time and Exam Score")
plt.legend()
plt.grid(color="gray", linestyle=":")
```

```python
# plt.scatter(x, y, color="color_name", marker="marker style", label="label_name")
# scatter plot
plt.scatter([1, 2, 3], [50, 55, 60], color="blue", marker="o", label="Class A")
plt.scatter([1, 2, 3], [45, 50, 52], color="orange", marker="o", label="Class B")
plt.xlabel("Hours Studied")
plt.ylabel("Exam Score")
plt.title("Comparison of Two Classes")
plt.legend()
plt.grid(color="gray", linestyle=":")
```
