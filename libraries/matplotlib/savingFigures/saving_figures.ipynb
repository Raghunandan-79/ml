```python
import matplotlib.pyplot as plt
```

```python
# savefig("filename.extension", dpi=value, bbox_inches="tight")
x = [1, 2, 3, 4]
y = [10, 20, 15, 25]

# create plot
plt.plot(x, y, color="blue", marker="o")
plt.title("Simple Line Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

plt.savefig("Line-Plot.png", dpi=300, bbox_inches="tight")
plt.show()
```
